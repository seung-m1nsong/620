## **6. Does the observed relative frequency of these words follow Zipf’s law? Explain.**

When Professor Ziff draws a rank-frequency graph of words in English newspapers, he shows that the distribution of words follows a power law distribution.

In other words, if the most used word was used 1,000 times, the second word would be used 500 times, the next word would be 333, 250... and so on!

What's special about the power law distribution is that when the scale of each axis is converted to log, it takes the shape of a straight line, and Professor Ziff showed that the slope of the rank-frequency distribution of words in English newspapers is 1.

The most frequent word must appear almost twice as often as the second most frequent word to be called Zipf's Law, but this is not the case in this text.

The frequency of the most frequently occurring word, The, is 3.27%, and the second, to, is 3.23%, so it is difficult to say that it follows Zipf’s Law.

```python
# Print the 200 highest frequency words, their counts, and their relative frequencies as percentages
for word, count in words_200:
    percentage = (count / words) * 100  # Calculate the percentage
    print(f"{word}: Count = {count}, Percentage = {percentage:.2f}%")
```
